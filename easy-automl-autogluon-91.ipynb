```python
id = test['id']
train.drop(['id'], axis=1, inplace=True)
test.drop(['id'], axis=1, inplace=True)
```

# 2. AutoML Modelling

```python
!pip install --upgrade autogluon
```

```python
from autogluon.tabular import TabularPredictor, TabularDataset

# Load the training dataset
train_data = TabularDataset(train)
test_data = TabularDataset(test)

# Define the target variable
label_column = "NObeyesdad"

# Define and train the AutoGluon model
automl = TabularPredictor(label=label_column, problem_type='multiclass', eval_metric='accuracy')
automl.fit(train_data)

# Make predictions on test_data
predictions = automl.predict(test_data)
print(predictions)
```

```python
leaderboard = automl.leaderboard()
leaderboard
```

# 3. Making Predictions and Submission File

```python
y_submission = pd.DataFrame(predictions, columns=['NObeyesdad'])
y_submission.head()
```

```python
submission_to_do = pd.concat([id, y_submission], axis = 1)
submission_to_do.to_csv('submission.csv', index = False)
```
